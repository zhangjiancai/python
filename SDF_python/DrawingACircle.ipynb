```python
import math,time
```

```python
def circle(x:float,y:float,c:float,d:float) -> float:
    radius = 0.5
    return math.sqrt((x-c)**2+(y-d)**2)-radius
```

```python
def sample(x:float,y:float,c:float,d:float) -> str:
    if circle(x,y,c,d)<=0:
        return '#'
    else:
        return '*' 
```

```python
def mainforthis(a:int ,b:int,c:int ,d:int):
  frame_chars = []
  remapped_c = c/b*2 -1 
  remapped_d = (d/a*2-1)*(2*a/b)

  for y in range(a):
    for x in range(b):
      # remap to -1..1 range (for x)...
      remapped_x = x / b * 2 - 1
      # ...and corrected for aspect ratio range (for y)
      remapped_y = (y / a * 2 - 1) * (2 * a/b)
      frame_chars.append(sample(remapped_x, remapped_y,remapped_c,remapped_d))
    frame_chars.append('\n')
  return frame_chars
  
```

```python
print ( ''.join(mainforthis(80,80,10,30)) )

```
